```python
import pathlib
import sys
sys.path.insert(0,'..')
import seqparser
from seqparser import (FastaParser, FastqParser, transcribe, reverse_transcribe)
#import pytest
def main(print_faCheck=True):
    data_dir = pathlib.Path.cwd().parent
    fasta_file = data_dir / "data/test.fa"
    
    # Create instance of FastaParser
    
    fasta_parser = FastaParser(fasta_file)
   
    # For each record of FastaParser, Transcribe the sequence
    # and print it to console

    faCheck = "-" * 50
    faCheck = "\n" + faCheck + "\n"

    fasta_file = fasta_file.name
    
    if print_faCheck:
        print(faCheck + f"Transcribing FASTA: {fasta_file}\n")

    
    for seq_name, seq in fasta_parser:
        with open ("test1.txt", "a") as f:
            print(seq_name, transcribe(seq), file=f)
       
            
        
```
